```python
import numpy as np
```

```python
def iou_loss(actual, predicted) -> float:
    return np.sum(np.abs(actual-predicted))
    
    
```

```python
x = np.random.rand(10,10,7)
y = np.random.rand(10,10,7)
```

```python
np.sum(np.abs(x-y))
```

```python
iou_loss(x,y)
```
